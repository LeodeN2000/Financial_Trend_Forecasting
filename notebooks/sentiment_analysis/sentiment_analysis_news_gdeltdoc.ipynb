```python
import pandas as pd
from datetime import datetime, timedelta, date
import time
from transformers import pipeline
```

```python
from gdeltdoc import GdeltDoc, Filters

f = Filters(
    keyword = "TESLA",
    start_date = "2023-11-26",
    end_date = "2023-11-27", 
    country = "US"
)

gd = GdeltDoc()

# Search for articles matching the filters
articles = gd.article_search(f)

# Get a timeline of the number of articles matching the filters
timeline = gd.timeline_search("timelinevol", f)
```

```python
columns = articles.columns
```

```python
def fetch_daily(time_window, stock): 
    df = pd.DataFrame(columns=columns)
    for d in range(time_window): 
        try:
            f = Filters(
            keyword = stock,
            start_date = (date.today() - timedelta(days=d)).strftime("%Y-%m-%d"),
            end_date = (date.today() - timedelta(days=d) + timedelta(days=1)).strftime("%Y-%m-%d"), 
            country = "US"
        )
        
            gd = GdeltDoc()
            
            # Search for articles matching the filters
            articles = gd.article_search(f)
            
            # Get a timeline of the number of articles matching the filters
            timeline = gd.timeline_search("timelinevol", f)
            df = pd.concat([df, articles])
            time.sleep(5)
        except KeyError as e:
            print(f"A KeyError occurred: {e} for step {d}")
    return df
```

```python
def classifier(df): 
    classifier = pipeline("text-classification", model = "roberta-large-mnli")
    is_ev_market = []
    is_tesla_stock = []
    for row in df.title[0:1000]:     
        is_ev_market.append(classifier(f"{row}. The headline is related to the Electric vehicle market."))
        is_tesla_stock.append(classifier(f"{row}. The headline is related to TESLA financial company's performance."))
        #print(f"The news is: {row}. The relation with EV market is: {is_ev_market}. The relation with the TESLA company is: {is_tesla_stock}")
    
    return is_ev_market, is_tesla_stock
        
```
